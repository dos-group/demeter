```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#  Read the CSV file
data = pd.read_csv("2_demeter_CARS_24_new/data/demeter-throughput-data.csv", sep="|")
#data = pd.read_csv("2_demeter_ADS_24_small_new/data/demeter-throughput-data.csv", sep="|")

# Reading and Preparing the demeter.csv Data
demeter_data = pd.read_csv("2_demeter_CARS_24_new/demeter.csv")
#demeter_data = pd.read_csv("2_demeter_ADS_24_small_new/demeter.csv")
```

```python
# Convert timestamp to seconds since start
data["seconds_since_start"] = data["timestamp"] - data["timestamp"].iloc[0]

# Step 3: Smooth the data using a rolling window average
window_size = 1200  # You can adjust this as needed
data["smoothed_value"] = data["value"].rolling(window=window_size).mean()

# Define common y-axis range
y_range = (data["value"].min(), data["value"].max())

# Visualization
fig, axs = plt.subplots(2, 1, figsize=(12,10))

# Original Data
axs[0].plot(data["seconds_since_start"], data["value"], label="Original Data", color="blue")
axs[0].set_title("Original Throughput Metrics Over Time")
axs[0].set_xlabel("Seconds Since Start")
axs[0].set_ylabel("Value")
axs[0].set_ylim(y_range)
axs[0].legend()

# Smoothed Data
axs[1].plot(data["seconds_since_start"], data["smoothed_value"], label="Smoothed Data", color="red")
axs[1].set_title("Smoothed Throughput Metrics Over Time")
axs[1].set_xlabel("Seconds Since Start")
axs[1].set_ylabel("Value")
axs[1].set_ylim(y_range)
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Calculate numNodes for each row in demeter_data
demeter_data["numNodes"] = demeter_data["scaleOut"] / demeter_data["taskSlots"]

# Calculate CPU and Memory usage for each row
demeter_data["cpu_usage"] = demeter_data["cpuVal"] * demeter_data["numNodes"]
demeter_data["memory_usage"] = demeter_data["memVal"] * demeter_data["numNodes"]

print(demeter_data)

# Initializing lists to store the CPU and Memory usage values for each second
cpu_values = []
memory_values = []

# Loop over all the seconds in the throughput dataset
for timestamp in data["timestamp"]:
    # Find the row in demeter_data where the timestamp falls between startTs of the current row and startTs of the next row
    current_row = demeter_data[(demeter_data["startTs"] <= timestamp) & (demeter_data["startTs"].shift(-1) > timestamp)]
    
    # If a matching row is found, append the CPU and Memory usage values to their respective lists
    if not current_row.empty:
        cpu_values.append(current_row["cpu_usage"].values[0])
        memory_values.append(current_row["memory_usage"].values[0])
    # If no matching row is found (for the last timestamp), use the last available values
    else:
        cpu_values.append(cpu_values[-1])
        memory_values.append(memory_values[-1])

# Add the calculated CPU and Memory usage values to the data dataframe
data["cpu"] = cpu_values
data["memory"] = memory_values

# Visualization
fig, axs = plt.subplots(2, 1, figsize=(12,10))

# CPU Utilization Over Time
axs[0].plot(data["seconds_since_start"], data["cpu"], label="CPU Utilization", color="green")
axs[0].set_title("CPU Utilization Over Time")
axs[0].set_xlabel("Seconds Since Start")
axs[0].set_ylabel("CPU")
axs[0].legend()

# Memory Usage Over Time
axs[1].plot(data["seconds_since_start"], data["memory"], label="Memory Usage", color="purple")
axs[1].set_title("Memory Usage Over Time")
axs[1].set_xlabel("Seconds Since Start")
axs[1].set_ylabel("Memory")
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python
total_cpu_usage = sum(data["cpu"])
total_memory_usage = sum(data["memory"])
print(f"Total CPU Usage for 18 hours: {total_cpu_usage}")
print(f"Total Memory Usage for 18 hours: {total_memory_usage}")
```

```python
# Define function to smooth data
def smooth_data(data, window_size=60):
    return data["value"].rolling(window=window_size).mean()

# Initialize dictionaries to store total CPU and Memory usages for each method
total_cpu = {}
total_memory = {}

# Dictionary of file paths with corresponding method names
#files_and_methods = {
    #"1_baseline_ADS_24_small/data/baseline-task_managers-data.csv": "Static",
    #"1_reactive35_ADS_24_small/data/reactive-task_managers-data.csv": "Reactive",
    #"3_operator35_ADS_24_small/data/operator-task_managers-data.csv": "DS2",
#}

files_and_methods = {
    "1_baseline_CARS_24/data/baseline-task_managers-data.csv": "Static",
    "1_reactive35_CARS_24/data/reactive-task_managers-data.csv": "Reactive",
    "1_operator35_CARS_24/data/operator-task_managers-data.csv": "DS2",
}

# Initialize a figure to visualize smoothed graphs
fig, axs = plt.subplots(len(files_and_methods), 2, figsize=(15, len(files_and_methods) * 5))

# Loop over each file path
for idx, (file_path, method_name) in enumerate(files_and_methods.items()):
    # Read the file
    data = pd.read_csv(file_path, sep="|")
    
    # Sort data by timestamp
    data = data.sort_values(by="timestamp")
    
    # Convert timestamp to seconds since start for visualization
    data["seconds_since_start"] = data["timestamp"] - data["timestamp"].iloc[0]
    
    # Smooth the data
    data["smoothed_value"] = smooth_data(data)
    
    # Calculate CPU and Memory usage
    data["cpu"] = data["smoothed_value"] * 1
    data["memory"] = data["smoothed_value"] * 4096
    
    # Plot smoothed CPU utilization
    axs[idx, 0].plot(data["seconds_since_start"], data["cpu"], label=f"CPU Utilization ({method_name})", color="green")
    axs[idx, 0].set_title(f"Smoothed CPU Utilization Over Time ({method_name})")
    axs[idx, 0].set_xlabel("Seconds Since Start")
    axs[idx, 0].set_ylabel("CPU")
    axs[idx, 0].legend()
    
    # Plot smoothed Memory usage
    axs[idx, 1].plot(data["seconds_since_start"], data["memory"], label=f"Memory Usage ({method_name})", color="purple")
    axs[idx, 1].set_title(f"Smoothed Memory Usage Over Time ({method_name})")
    axs[idx, 1].set_xlabel("Seconds Since Start")
    axs[idx, 1].set_ylabel("Memory")
    axs[idx, 1].legend()

    # Sum the total CPU and Memory usage and store in dictionaries
    total_cpu[method_name] = data["cpu"].sum()
    total_memory[method_name] = data["memory"].sum()

# Adjust the layout
plt.tight_layout()
plt.show()

# Display the results
print("Total CPU Usage for Each Method:")
for method, cpu_usage in total_cpu.items():
    print(f"{method}: {cpu_usage}")

print("\nTotal Memory Usage for Each Method:")
for method, memory_usage in total_memory.items():
    print(f"{method}: {memory_usage}")
```

```python
# Adding the values for the "Demeter" method to the dictionaries
total_cpu["Demeter"] = total_cpu_usage
total_memory["Demeter"] = total_memory_usage

# Calculate the maximum CPU and Memory usage values for normalization
max_cpu = max(total_cpu.values())
max_memory = max(total_memory.values())

# Normalize the CPU and Memory usage values
normalized_cpu = {method: (cpu_usage / max_cpu) * 100 for method, cpu_usage in total_cpu.items()}
normalized_memory = {method: (memory_usage / max_memory) * 100 for method, memory_usage in total_memory.items()}

# Plotting
#labels = list(normalized_cpu.keys())
labels = list(["Static Configuration", "Flink Reactive", "DS2 Autoscaler", "Demeter"])
cpu_values = list(normalized_cpu.values())
memory_values = list(normalized_memory.values())

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(10,4))
rects1 = ax.bar(x - width/2, cpu_values, width, label='CPU', color="blue")
rects2 = ax.bar(x + width/2, memory_values, width, label='Memory', color="orange")

# Add some text for labels, title, and custom x-axis tick labels
ax.set_ylabel('Normalized Usage (%)')
ax.set_title('Normalized Resource Utilization for CPU and Memory')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Autolabel function for displaying the label on top of the bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(round(height, 2)),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.savefig("CARS_resource_utilization_2.pdf", bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Define font sizes
title_fontsize = 18
label_fontsize = 14
tick_fontsize = 16
annotation_fontsize = 14
y_axis_label_fontsize = 16

# Normalized CPU and Memory values (as percentages)
#cpu_values = [1.0, 0.6, 0.86, 0.81]  # Example values
#memory_values = [1.0, 0.6, 0.86, 0.63]  # Example values
#additional_cpu_percentage = 11.7 / 100  # 11.7%
#additional_memory_percentage = 5.4 / 100  # 5.4%

cpu_values = [1.0, 0.83, 0.9, 0.8]  # Example values
memory_values = [1.0, 0.83, 0.9, 0.57]  # Example values
additional_cpu_percentage = 16.6 / 100  # 16.6 % 
additional_memory_percentage = 8.2 / 100  # 8.2% 

# Labels for the methods
labels = ["Static Configuration", "Flink Reactive", "DS2 Autoscaler", "Demeter"]
x = np.arange(len(labels))
demeter_index = labels.index("Demeter")
width = 0.35

# Set up Seaborn styling
sns.set(style="whitegrid", palette="muted")
blue, yellow = sns.color_palette()[:2]
light_blue = sns.light_palette("blue", n_colors=3)[1]
light_yellow = sns.light_palette("orange", n_colors=3)[1]

# Function for autolabeling the bars
def autolabel(rects, additional=False):
    for rect in rects:
        height = rect.get_height()
        if additional:
            # Position for the additional bars' labels
            label_position = rect.get_y() + height
        else:
            # Position for the original bars' labels
            label_position = height
        ax.annotate('{}'.format(round(height, 2)),
                    xy=(rect.get_x() + rect.get_width() / 2, label_position),
                    xytext=(0, 3 if additional else -20),  # Adjust text position
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=annotation_fontsize)

# Creating the plot
fig, ax = plt.subplots(figsize=(10, 4))

# Plotting original bars
rects1 = ax.bar(x - width/2, cpu_values, width, label='CPU', color=blue)
rects2 = ax.bar(x + width/2, memory_values, width, label='Memory', color=yellow)

# Plotting the additional stacked bars for Demeter
rects1_add = ax.bar(demeter_index - width/2, additional_cpu_percentage, width, 
                    bottom=cpu_values[demeter_index], color=light_blue, label='Profiling CPU')
rects2_add = ax.bar(demeter_index + width/2, additional_memory_percentage, width, 
                    bottom=memory_values[demeter_index], color=light_yellow, label='Profiling Memory')

# Plotting additional stacked bars for Demeter
demeter_index = labels.index("Demeter")  # Index of Demeter in the labels list
rects1_add = ax.bar(demeter_index - width/2, additional_cpu_percentage, width, 
                    bottom=cpu_values[demeter_index], color=light_blue)
rects2_add = ax.bar(demeter_index + width/2, additional_memory_percentage, width, 
                    bottom=memory_values[demeter_index], color=light_yellow)

# Adding labels and title
ax.set_ylabel('Normalized Usage', fontsize=y_axis_label_fontsize)
ax.set_title('Normalized Resource Usage for CPU and Memory', fontsize=title_fontsize)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=tick_fontsize)
ax.legend(loc='lower left', framealpha=1, fontsize=label_fontsize)

# Autolabeling the bars
autolabel(rects1)
autolabel(rects2)
autolabel(rects1_add, additional=True)
autolabel(rects2_add, additional=True)

# Adjusting the y-axis limit
ax.set_ylim(0, 1.1)  # Up to a little above 1.0 for label clarity

fig.tight_layout()
plt.savefig("CARS_resource_usage.pdf", bbox_inches='tight')
plt.show()
```
