```python
# Create individual graphs for each workload range, only plotting non-empty ranges
import re
import numpy as np

# Extract the dictionary part
chunk_dict_str = re.search(r"Prepared Chunk-Dictionary: (\{.*\})", data_str).group(1)
chunk_dict = eval(chunk_dict_str)

# Count non-empty ranges to adjust figure size accordingly
non_empty_count = sum([1 for range_tuple, count in chunk_dict.items() if not data["smoothed_value"].between(range_tuple[0], range_tuple[1]).sum() == 0])

plt.figure(figsize=(12, 5 * non_empty_count))

subplot_index = 1
for i, (range_tuple, count) in enumerate(reversed(list(chunk_dict.items()))):
    
    # Create a series with the same length as the smoothed data
    range_series = data["smoothed_value"].copy()
    
    # Set values outside the current range to NaN
    range_series[(range_series < range_tuple[0]) | (range_series > range_tuple[1])] = np.nan
    
    # Only plot non-empty ranges
    if not range_series.isna().all():
        ax = plt.subplot(non_empty_count, 1, subplot_index)
        subplot_index += 1

        ax.plot(data["seconds_since_start"], range_series, label=f"Workload: {range_tuple[0]} to {range_tuple[1]}", color="red")
        ax.set_title(f"Workload Range: {range_tuple[0]} to {range_tuple[1]}")
        ax.set_xlabel("Seconds Since Start")
        ax.set_ylabel("Value")
        ax.set_xlim([data["seconds_since_start"].min(), data["seconds_since_start"].max()])  # Keeping x-axis consistent
        ax.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Checking the first few rows of the data to understand its structure
demeter_data.head()

# Helper function to find the matching configuration for a given timestamp
def find_matching_config(timestamp, demeter_data):
    matching_config = demeter_data[demeter_data['startTs'] <= timestamp].iloc[-1]
    return matching_config

# Visualization: Plotting combined normalized CPU and Memory datasets for each workload range in descending order of the workload range
sorted_ranges = sorted(chunk_dict.keys(), key=lambda x: x[0], reverse=True)

for range_tuple in sorted_ranges:
    # Extract corresponding datasets
    cpu_values = np.array(cpu_datasets.get(range_tuple, [])) / 24  # Normalizing
    mem_values = np.array(memory_datasets.get(range_tuple, [])) / (24 * 4096)  # Normalizing
    
    # Only plot if there's data
    if cpu_values.size and mem_values.size:
        fig, ax1 = plt.subplots(figsize=(12, 5))
        
        # Plot CPU values lightly
        ax1.plot(cpu_values, color="blue", alpha=0.3, linestyle='-', label="CPU Resource Value")
        ax1.set_xlabel("Seconds in Workload Range")
        ax1.set_ylabel("CPU Utilization", color="blue")
        ax1.tick_params(axis='y', labelcolor="blue")
        
        # Compute and plot CPU regression line
        x = np.arange(len(cpu_values))
        m, b = np.polyfit(x, cpu_values, 1)
        ax1.plot(x, m*x + b, color="blue", linestyle='-', label="CPU Regression")
        
        # Create a second y-axis for Memory values
        ax2 = ax1.twinx()
        # Plot Memory values lightly
        ax2.plot(mem_values, color="red", alpha=0.3, linestyle='--', label="Memory Resource Value")
        ax2.set_ylabel("Memory Utilization", color="red")
        ax2.tick_params(axis='y', labelcolor="red")
        
        # Compute and plot Memory regression line
        m, b = np.polyfit(x, mem_values, 1)
        ax2.plot(x, m*x + b, color="red", linestyle='--', label="Memory Regression")
        
        # Set the y-limits for normalization
        ax1.set_ylim(0, 1)
        ax2.set_ylim(0, 1)
        
        # Combining legends from both ax1 and ax2
        lines, labels = ax1.get_legend_handles_labels()
        lines2, labels2 = ax2.get_legend_handles_labels()
        ax2.legend(lines + lines2, labels + labels2, loc="lower left")
        
        fig.tight_layout()
        plt.title(f"Normalized CPU and Memory Values with Regression for Workload Range: {range_tuple[0]} to {range_tuple[1]}")
        plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the color palette
sns.set_palette("muted")
cpu_color, mem_color = sns.color_palette()[:2]

# Set font sizes
title_fontsize = 16
label_fontsize = 14
tick_fontsize = 12

# Visualization: Plotting combined normalized CPU and Memory datasets for each workload range in descending order of the workload range
sorted_ranges = sorted(chunk_dict.keys(), key=lambda x: x[0], reverse=True)

for range_tuple in sorted_ranges:
    # Extract corresponding datasets
    cpu_values = np.array(cpu_datasets.get(range_tuple, [])) / 24  # Normalizing
    mem_values = np.array(memory_datasets.get(range_tuple, [])) / (24 * 4096)  # Normalizing
    
    # Only plot if there's data
    if cpu_values.size and mem_values.size:
        fig, ax1 = plt.subplots(figsize=(10, 5))
        
        # Plot CPU values lightly
        ax1.plot(cpu_values, color=cpu_color, alpha=0.3, linestyle='-', label="CPU Resource Value")
        ax1.set_xlabel("Seconds in Workload Segement", fontsize=label_fontsize)
        ax1.set_ylabel("CPU Utilization", color=cpu_color, fontsize=label_fontsize)
        ax1.tick_params(axis='y', labelcolor=cpu_color, labelsize=tick_fontsize)
        
        # Compute and plot CPU regression line
        x = np.arange(len(cpu_values))
        m, b = np.polyfit(x, cpu_values, 1)
        ax1.plot(x, m*x + b, color=cpu_color, linestyle='-', label="CPU Regression")
        
        # Create a second y-axis for Memory values
        ax2 = ax1.twinx()
        # Plot Memory values lightly
        ax2.plot(mem_values, color=mem_color, alpha=0.3, linestyle='--', label="Memory Resource Value")
        ax2.set_ylabel("Memory Utilization", color=mem_color, fontsize=label_fontsize)
        ax2.tick_params(axis='y', labelcolor=mem_color, labelsize=tick_fontsize)
        
        # Compute and plot Memory regression line
        m, b = np.polyfit(x, mem_values, 1)
        ax2.plot(x, m*x + b, color=mem_color, linestyle='--', label="Memory Regression")
        
        # Set the y-limits for normalization
        ax1.set_ylim(0, 1)
        ax2.set_ylim(0, 1)
        
        # Combining legends from both ax1 and ax2
        lines, labels = ax1.get_legend_handles_labels()
        lines2, labels2 = ax2.get_legend_handles_labels()
        ax2.legend(lines + lines2, labels + labels2, loc="lower left", fontsize=label_fontsize)
        
        fig.tight_layout()
        plt.title(f"Normalized CPU and Memory with Regression for Workload Segement: {range_tuple[0]} to {range_tuple[1]}", fontsize=title_fontsize)
        plt.show()
```
